## Problem Statement:
Predict the percentage of an student based on the no. of study hours. This is a simple linear regression task as it involves just 2 variables.

###  Step1:-- Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import lux
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Step2:-- Importing the Data-Set

```python
url = "http://bit.ly/w-data"
s_data = pd.read_csv(url)
print("The Dataset ")

s_data.head(10)
```

### Step3-- Data Visualization 

```python
s_data.head(25)
```

### Step:4--Preparing Data 

```python
X = s_data.iloc[:, :-1].values  
Y = s_data.iloc[:, 1].values  
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state =0)
```

### Step5- Training the Model

```python
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
print("Data Trained Sucessfully")
```

### Step:6--Regression Line 

```python
slope = regressor.coef_
interc = regressor.intercept_
print("Slope:",slope,"\nY-Intercept:",interc)
```

```python
## Plotting the Trained Regression Model
Relation = slope*X + interc
plt.scatter(X,Y)
plt.plot(X,Relation)
plt.show
```

### Step:7-Making Prediction 
#### by Using Trained Data 

```python
y_pred = regressor.predict(X_test)
print("Hours Studied\n", X_test)
df1 = pd.DataFrame({'Actual score':Y_test, 'Predicted score': y_pred})
df1
```

```python
df1.plot(kind = 'line')
plt.title("Graph for Actual Percentage vs Preicted Percentage")
plt.xlabel("Student No.-->")
plt.ylabel("Percentage score-->")
plt.show()
```

### The Results

```python
hours = 9.25
fin_pred = round(regressor.predict([[9.25]])  [0],2)
print("No of Hours studied = {}".format(hours))
print("Predicted Score = {}".format(fin_pred))
```

```python
from sklearn import metrics
print('Mean Absolute Error: ', metrics.mean_absolute_error(Y_test,y_pred))
```
